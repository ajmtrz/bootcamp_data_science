Now we will create 4 lists with all the players of both teams, 2 list with the players that began the match of both teams and other 2 with the players that enter during the match

```python
start_p_h = [n for n in f_h[f_h.period == 2].sort_values('id_pos').trackable_object.unique() if n in f_h[f_h.period == 1].trackable_object.unique()]
subs_h = [n for n in f_h[f_h.period == 2].sort_values('id_pos').trackable_object.unique() if n not in f_h[f_h.period == 1].trackable_object.unique()]
start_p_a = [n for n in f_a[f_a.period == 2].sort_values('id_pos').trackable_object.unique() if n in f_a[f_a.period == 1].trackable_object.unique()]
subs_a = [n for n in f_a[f_a.period == 2].sort_values('id_pos').trackable_object.unique() if n not in f_a[f_a.period == 1].trackable_object.unique()]
```

Now we will use this lists for plot an individual heatmap for each player and classify it in a tactical line

**Players of Home Team**

```python
for p in start_p_h:
    pl = PreprocessData.generate_player_match_data('data/matches/2841/structured_data.json','data/matches/2841/match_data.json', p)
    PlotGraphs.plot_player_heatmap(pl)
```

**Players Subtitutions of Home Team**

```python
for p in subs_h:
    pl = PreprocessData.generate_player_match_data('data/matches/2841/structured_data.json','data/matches/2841/match_data.json', p)
    PlotGraphs.plot_player_heatmap(pl)
```

**Players of Home Team**

```python
for p in start_p_a:
    pl = PreprocessData.generate_player_match_data('data/matches/2841/structured_data.json','data/matches/2841/match_data.json', p)
    PlotGraphs.plot_player_heatmap(pl)
```

**Players Subtitutions of Home Team**

```python
for p in subs_a:
    pl = PreprocessData.generate_player_match_data('data/matches/2841/structured_data.json','data/matches/2841/match_data.json', p)
    PlotGraphs.plot_player_heatmap(pl)
```
