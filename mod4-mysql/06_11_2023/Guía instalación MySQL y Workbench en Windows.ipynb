Finalizada la instalacin de los mdulos ser turno de proceder a una configuracin inicial antes de ejecutar los correspondientes servicios. Pulsamos Next y elegimos la primera opcin Standalone MySQL Server/Classic MySQL Replication



La siguiente pantalla es importante, ya que necesitaremos configurar algunos parmetros como el tipo de equipo que tendremos para SQL, adems de protocolos y puertos TCP por donde se efectuarn las conexiones remotas al servidor SQL.

Para la configuracin de tipo de ordenador tendremos tres opciones distintas:

`Server Computer:` esta segunda opcin ser orientada a ordenadores utilizados para funciones de servidor, por ejemplo, servidor web con bases de datos.

`Dedicated Computer:` la tercera opcin es par el caso en que queremos crear un equipo solo y exclusivamente orientado a bases de datos. Por ejemplo, una mquina virtual en la que se almacenen nuestras bases de datos.
La siguiente opcin que tendremos que elegir es la del puerto TCP que utilizaremos para conexiones remotas. Por defecto es el 3306. La opcin que marquemos aqu ser el puerto que tendremos que abrir en nuestro router para establecer las conexiones remotas.

El resto de opciones recomendamos dejarlas por defectos tal y como estn.

Finalmente configuraremos el nombre del servicio para MySQL y las preferencias generales en cuando a inicio del demonio y el uso de cuentas de usuario.



Para finalizar, en la ltima pantalla pulsamos en Execute para ejecutar las acciones y activar los servicios correspondientes en el sistema. Todo debera de haberse completado correctamente.
