### Ejercicio 5:
- Haz el producto punto de los siguientes vectores y encuentra el angulo entre los vectores.

$\overrightarrow{AB} = 3\widehat{i} + 4\widehat{j} $


$\overrightarrow{CD} = -2\widehat{i} + 7.5\widehat{j} $
 

```python
AB = (3, 4)
CD = (-2, 7.5)

producto_escalar = np.dot(AB, CD)
producto_escalar
```

```python
modAB = np.linalg.norm(AB)
modCD = np.linalg.norm(CD)
print(f"{modAB}, {modCD}")
```

```python
# Radianes

alpha = np.arccos(producto_escalar / (modAB * modCD))
alpha
```

```python
# Grados

np.degrees(alpha)
```
