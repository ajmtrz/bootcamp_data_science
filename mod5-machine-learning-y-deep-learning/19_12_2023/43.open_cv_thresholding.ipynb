```python
import numpy as np
import matplotlib.pyplot as plt

import cv2
```

### Imagen Thresholding

La umbralizacin (thresholding) es el mtodo ms simple de la segmentacin de imgenes cuyo objetivo es separar un objeto de inters del fondo de una imagen. Hay que tomar en cuenta que para su aplicacin deben emplearse imgenes en las que el objeto y el fondo sean diferenciables, ya que si no lo son, va a ser mucho ms difcil determinar dicho objeto.



```python
img_thresh = cv2.imread("../Data/ejemplo_thresholding.jpg")

# Para aplicar thresholding es recomendable transformar la imagen a escala de grises

img_thresh = cv2.cvtColor(src = img_thresh, code = cv2.COLOR_BGR2GRAY)

plt.imshow(img_thresh, cmap = "gray")
plt.show()
```

```python
img_thresh = cv2.imread("../Data/ejemplo_thresholding.jpg")

# Para aplicar thresholding es recomendable transformar la imagen a escala de grises

img_thresh = cv2.cvtColor(src = img_thresh, code = cv2.COLOR_BGR2RGB)

plt.imshow(img_thresh, cmap = "gray")
plt.show()
```

### Thresholding Binario

Transforma los numero de la imagen a 0 o 1 dependiendo si el pixel es mayor o menor a el valor de **`thresh`**.

```python
thresh, img_thresh = cv2.threshold(src = img_thresh, thresh = 127, maxval = 255, type = cv2.THRESH_BINARY)

plt.imshow(img_thresh, cmap = "gray")
plt.show()
```

### Thresholding Truncado

Transforma los numeros de la imagen al valor de **`tresh`** si son mayores a **`thresh`**, de lo contrario no los modifica. Se ignora **`maxval`**.

```python
# Imagen Original

img_thresh = cv2.imread("../Data/ejemplo_thresholding.jpg")

img_thresh = cv2.cvtColor(src = img_thresh, code = cv2.COLOR_BGR2GRAY)
```

```python
thresh, img_thresh = cv2.threshold(src = img_thresh, thresh = 127, maxval = 255, type = cv2.THRESH_TRUNC)

plt.imshow(img_thresh, cmap = "gray")
plt.show()
```

### Ejemplo de aplicacion

Queremos separar el texto de la siguiente imagen del fondo gris.

```python
# Usando flags = 0, leemos la imagen en escala de grises en lugar de BGR

img = cv2.imread("../Data/threshholding_crossword.jpg", flags = 0)

plt.figure(figsize = (10, 10))
plt.imshow(img, cmap = "gray")
plt.show()
```

### Thresholding Binario

```python
thresh, img_1 = cv2.threshold(src    = img,
                              thresh = 127,
                              maxval = 255,
                              type   = cv2.THRESH_BINARY)

plt.figure(figsize = (10, 10))
plt.imshow(img_1, cmap = "gray")
plt.show()
```

### Adaptive Threshold

- **`blockSize`**: Tamao de la "vecindad" de pxeles que se utiliza para calcular el **`threshold`**.
- **`C`**: Constante restada de la media o media ponderada.

```python
img_2 = cv2.adaptiveThreshold(src            = img,
                              maxValue       = 255,
                              adaptiveMethod = cv2.ADAPTIVE_THRESH_MEAN_C,
                              thresholdType  = cv2.THRESH_BINARY,
                              blockSize      = 11,
                              C              = 8)

plt.figure(figsize = (10, 10))
plt.imshow(img_2, cmap = "gray")
plt.show()
```

```python
img_3 = cv2.adaptiveThreshold(src            = img,
                              maxValue       = 255,
                              adaptiveMethod = cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                              thresholdType  = cv2.THRESH_BINARY,
                              blockSize      = 11,
                              C              = 8)

plt.figure(figsize = (10, 10))
plt.imshow(img_3, cmap = "gray")
plt.show()
```

### Union: img_1, img_2

```python
union = cv2.addWeighted(src1 = img_1, alpha = 0.7, src2 = img_2, beta = 0.3, gamma = 0)

plt.figure(figsize = (10, 10))
plt.imshow(union, cmap = "gray")
plt.show()
```
