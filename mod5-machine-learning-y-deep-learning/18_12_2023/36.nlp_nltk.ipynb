## Natural Language Processing (NLP)

El procesamiento del lenguaje natural (**NLP**), se define ampliamente como la manipulación automática del lenguaje natural, como el habla y el texto, por parte del software.

El estudio de **NLP** existe desde hace más de 50 años y surgió del campo de la lingüística con el auge de los ordenadores.

**El lenguaje natural se refiere a la forma en que nosotros, los humanos, nos comunicamos entre nosotros.**

Dada la importancia de este tipo de datos, debemos tener métodos para comprender y razonar sobre el lenguaje natural, tal como lo hacemos con otros tipos de datos.

En Python, tenemos la librería **NLTK** (**Natural Language ToolKit**), éste es un modulo que contiene herramientas para el manejo del lenguaje natural.

### ¿Qué podemos hacer con NLTK?

Con NLTK, podemos realizar una variedad de tareas de PNL, como:

   - `Análisis de sentimientos`: Determinar el sentimiento de un texto, como positivo, negativo o neutral.
   - `Partición del texto`: Dividir el texto en sus componentes, como palabras, frases o oraciones.
   - `Traducción`: Traducir texto de un idioma a otro.
   - `Generación de texto`: Generar texto nuevo, como un resumen de un artículo o una historia.

#### Componentes principales de NLTK

   - `Corpus`: Un corpus es una colección de textos, normalmente etiquetados, que se utilizan para entrenar modelos de procesamiento del lenguaje natural (PLN). Los corpus pueden ser de diferentes tipos, como corpus de noticias, corpus de literatura o corpus de código. Los corpus etiquetados contienen información adicional sobre cada texto, como el género, el tema o la fecha de publicación.
   - `Tokenización`: La tokenización es el proceso de dividir un texto en sus componentes básicos, como palabras, frases u oraciones. La tokenización es un paso fundamental en muchas tareas de PLN, como el análisis sintáctico, la búsqueda de información y el análisis de sentimientos.
   - `Lematización`: La lematización es el proceso de convertir una palabra a su forma léxica básica. Esto es importante para tareas de PLN como el análisis sintáctico y la búsqueda de información. La lematización se basa en el concepto de lema, que es la forma base de una palabra, independientemente de su forma flexionada.
   - `Análisis sintáctico`: El análisis sintáctico es el proceso de determinar la estructura sintáctica de un texto. Esto permite comprender cómo se relacionan las palabras entre sí y cómo se puede interpretar el texto. El análisis sintáctico se basa en la teoría de la gramática generativa, que establece que las oraciones se pueden generar a partir de un conjunto de reglas gramaticales.
   - `Relación semántica`: La relación semántica es la relación entre dos palabras. Las palabras pueden estar relacionadas por significado, forma o contexto. La relación semántica se puede utilizar para tareas de PLN como la búsqueda de información, la traducción automática y el análisis de sentimientos.

### Riqueza Léxica

La riqueza léxica es la relación que existe entre la extensión de un texto y el número de palabras distintas que contiene. 

````python
riqueza_lexica = palabras_diferentes / total_palabras

print(f"Riqueza Lexica: {riqueza_lexica}")
````

````python
# Funcion para calcular la riqueza lexica

def riqueza_lexica_fun(tokens):
    
    total_palabras = len(tokens)
    palabras_diferentes = len(set(tokens))
    
    riqueza_lexica = palabras_diferentes / total_palabras
    
    return riqueza_lexica
````

````python
riqueza_lexica_fun(tokens)
````

### .text.ContextIndex() y .similar_words()

**`.text.ContextIndex`** y **`.similar_words`** son utilizados para encontrar palabras similares.

**Simililares no significa que sean sinónimos, sino que son palabras que van a encontrarse en un contexto similar o tienen similitud con la palabra de busqueda.**

````python
# Si quisieramos encontrar las palabras similares de más de una palabra podemos usar el siguiente bucle:

similares = list()

for word in nltk.word_tokenize("naturaleza madre"):
    similares.append(nltk.text.ContextIndex(tokens).similar_words(word))
    
pd.DataFrame(data = similares, index = ["naturaleza", "madre"]).T
````

### STOPWORDS

Las **`stopwords`** (palabras funcionales o palabras vacias) son las palabras sin significado como artículos, pronombres, preposiciones, etc. que son filtradas antes o después del procesamiento de datos **NLP**.

````python
stopwords = nltk.corpus.stopwords.words("spanish")

for stopword in stopwords:
    print(stopword, end = " | ")
````

````python
tokens = nltk.word_tokenize(text = texto, language = "spanish")

tokens_limpios = list() 

tokens = [token.lower( ) for token in tokens]

for token in tokens: 
    if token not in stopwords: 
        tokens_limpios.append(token)
        
print(f"Tamaño original: {len(tokens)}") 
print(f"Tamaño despues de stopwords: {len(tokens_limpios)}")
````

````python
nltk.FreqDist(nltk.Text(tokens_limpios)).plot(10)

plt.show()
````

````python
# Eliminamos elementos de tamaño 2 o menor:

tokens_limpios = list()

for token in tokens: 
    if token not in stopwords: 
        if len(token) > 2 and token != '...': 
            tokens_limpios.append(token)
            
print(len(tokens_limpios))
````

````python
nltk.FreqDist(nltk.Text(tokens_limpios)).plot(20)

plt.show()
````

### Stemming
 
Para hacer stemming usamos el algoritmo _**Porter Stemmer:** https://tartarus.org/martin/PorterStemmer/_

#### ¿En que consiste el algoritmo de `Poter Stemmer`?

El algoritmo de Porter Stemmer es un algoritmo simple pero eficiente de stemming que funciona aplicando una serie de reglas a las palabras en inglés. Estas reglas eliminan los sufijos comunes de las palabras, dejando solo la raíz de la palabra.

**Funciona en tres pasos**:

   - `Eliminación de sufijos comunes`: El algoritmo comienza eliminando los sufijos comunes de las palabras. Los sufijos comunes incluyen:
        -ing
        -ed
        -es
        -ly
        -s
        
   - `Eliminación de sufijos irregulares`: El algoritmo luego elimina los sufijos irregulares que no se eliminan en el primer paso. Algunos sufijos irregulares incluyen:
        -en
        -er
        -est
        -ize
        
   - `Remoción de sufijos de verbos irregulares`: El algoritmo luego elimina los sufijos de verbos irregulares. Algunos sufijos de verbos irregulares incluyen:
        -a
        -e
        -i
        -o
        
Ejemplos de cómo funciona el algoritmo de Porter Stemmer:

- running → run
- walked → walk
- horses → hors
- beautiful → beaut
- loving → love
- singed → sing
- learned → learn
- be → be

El algoritmo de Porter Stemmer tiene algunas limitaciones. Por ejemplo, no puede manejar palabras compuestas o palabras que tienen más de un significado. Sin embargo, es un algoritmo eficaz para la mayoría de las tareas de PLN.

````python
from nltk.stem import PorterStemmer
````

````python
texto = "hi dear students, how are you doing? we are almost finishing the bootcamp. we hope you are learning and enjoying the course."
texto
````

````python
# Inicializamos un objeto PorterStemmer()
stemmer = PorterStemmer()

" ".join([stemmer.stem(word) for word in  nltk.word_tokenize(text = texto, language = "english")])
````

````python
stemmer.stem('learning')
````

### Lematización
Similar al stemming, pero reduce la palabra a una raíz que sí tiene significado:
+ going, goes, gone --> go
+ intelligence, intelligently --> intelligent

````python
from nltk.stem import WordNetLemmatizer
````

````python
texto = "hi dear students, how are you doing? we are almost finishing the bootcamp. we hope you are learning and enjoying the course."
texto
````

````python
# Inicializamos un objeto WordNetLemmatizer()
lemmatizer = WordNetLemmatizer()

" ".join([lemmatizer.lemmatize(word, pos="v") for word in  nltk.word_tokenize(text = texto, language = "english")])
````

### Analisis de Sentimiento

````python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()
````

````python
texto = """The National Guard has been released in Minneapolis to do the job that the Democrat Mayor couldn’t do.
           Should have been used 2 days ago & there would not have been damage & Police Headquarters would not have
           been taken over & ruined. Great job by the National Guard. No games!"""

tokens = nltk.word_tokenize(text = texto)
````

````python
for token in tokens:
    ss = sia.polarity_scores(token)
    print(token)
    print(ss)
````
