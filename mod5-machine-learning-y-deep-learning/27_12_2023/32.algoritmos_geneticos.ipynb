## Algoritmos Genticos

Un algoritmo gentico es un algoritmo heurstico basado en la teora de la evolucin de las especies.

- Para resolver problemas de optimizacin y bsqueda. Estos algoritmos son de suma importancia por su efectividad comprobada para tratar de resolver problemas de complejidad P y NP.


- Los algoritmos genticos son parte de la inteligencia artificial y es la base de los algoritmos bio-inspirados y evolutivos. Los cuales son aplicados en diferentes reas del conocimiento para resolver problemas no determinsticos.

### Evolucin Natural
En la naturaleza los procesos evolutivos ocurren cuando se satisfacen las siguientes condiciones:
- Una entidad o individuo tiene la habilidad de reproducirse.
- Hay una poblacin de tales individuos que son capaces de reproducirse.
- Existe alguna variedad, diferencia, entre los individuos que se reproducen.
- Algunas diferencias en la habilidad para sobrevivir en el entorno estn asociadas con esa diversidad entre los individuos.



Un Algoritmo Gentico puede ser visto como una estructura de control que organiza o dirige un conjunto de transformaciones y operaciones diseadas para simular procesos de evolucin.

Tambin se pueden considerar como mtodos de bsqueda aleatoriamente guiados hacia un objetivo o varios objetivos con ciertas restricciones y a la vez libertades.

### Procesos y Componentes del Algoritmo Gentico

### 1. **Inicializacin de la Poblacin**:
- La poblacin es un conjunto de soluciones candidatas y cada una de las soluciones es un individuo representado por un cromosoma.
- La inicializacin se puede realizar:
    - Uniforme sobre el espacio de bsqueda.
    - Cadena binaria con probabilidad de 0.5 aplicada a cada gen.
    - Representacin real: uniforme sobre un intervalo dado.

- Tambin se puede elegir la poblacin a partir de los resultados de una heurstica previa.


### 2. **Funcin de Evaluacin**:
- Es una medida que permite determinar cun buena es una solucin o individuo.
- Tiene como entrada el cromosoma que describe al individuo y generalmente devuelve un nmero real.
- Cuando hay restricciones, stas se pueden introducir en la funcin de evaluacin como penalizacin del costo.
    
    

### 3. **Parmetros**:
- Tamao de la Poblacin: Indica la cantidad de soluciones que forman la poblacin.
- Nmero de generaciones: Indica la cantidad de iteraciones que realizara el algoritmo hasta alcanzar una solucin satisfactoria.
- Probabilidad de cruce: Valor entre 0 y 1 que permite determinar si procede o no el cruce (con frecuencia el cruce es alto: 0.7 - 0.9).
- Probabilidad de Mutacin: Valor entre 0 y 1 que permite determinar si procede o no la mutacin (con frecuencia el valor de mutacin es pequeo: 0.01 - 0.15).
    

### 4. **Operadores**:
- **Operador de seleccin o Darwiniano**: Realiza la seleccin de los individuos de acuerdo a su adaptabilidad para el posterior cruce.
- **Operador de cruzamiento o Mendeliano**: Realiza la recombinacin del material gentico de dos individuos padres.
- **Operador de Mutacin**: Realiza la mutacin de un gen dentro de un cromosoma o individuo.

Para cada uno de estos operadores est asociado el uso de probabilidades y la generacin de nmeros aleatorios.
     
     
### 5. **Seleccin**:

Se debe de garantizar que los mejores individuos tengan una mayor posibilidad de ser padres (reproducirse) frente a individuos con bajo rendimiento. Sin embargo tambin se les tiene que dar la oportunidad de reproducirse a los individuos con bajo rendimiento porque puede que estos incluyan en su material gentico, configuraciones tiles en el proceso de reproduccin.

- **Seleccin por Torneo**: Para cada padre a seleccionar:
    - Escoger aleatoriamente k-individuos, seleccionar al mejor de ellos. 
    - "k" denomina el tamao del torneo, a mayor "k", mayor presin selectiva y viceversa.

    

- **Seleccin por Ruleta**:
    - Se le asigna una probabilidad de seleccin proporcional al valor del fitness a cada individuo.
    - Tiene una nica flecha que se mueve **`rand()*2pi`** en cada seleccin, si cae en al regin p1, se selecciona al individuo que genero esa regin.

- **Muestreo aleatorio universal**:
    - Tiene tantas flechas como individuos se desean seleccionar, con un ngulo entre ellas de 2pi/popsize. 
    - Las flechas se mueven **`rand()*2pi/ popsize`**.
    
    

### 6. **Cruzamiento**:
- Depende de la representacin usada (binaria, real o secuencial)
- Los hijos deberan heredar algunas caractersticas de cada padre.
- La recombinacin debe de producir cromosomas (o individuos validos).
- Se utiliza una probabilidad alta de actuacin sobre cada pareja de padres a cruzar.

- **Cruzamiento Binario**:
    -  Cada cromosoma se corta en 2 secciones, el corte se genera aleatoriamente entre 1 y el largo del cromosoma.

    

- **Cruzamiento Binario en varios puntos**:
    - Cada cromosoma se corta en n partes que son recombinadas.
    
    

- **Cruzamiento Binario Uniforme**:
    - Se escoge si el gen i-simo del hijo se toma del primer o del segundo padre, as sucesivamente.

- **Cruzamiento de Orden (Position-Based Crossover)**:
1. Seleccionar (al azar) un conjunto de posiciones del padre 1 (P1).
2. Producir un hijo borrando de P1 todos los valores, excepto aqullos que hayan sido seleccionados en el paso anterior.
3. Borrar los valores seleccionados del padre 2 (P2). La secuencia resultante de valores se usar para completar el hijo.
4. Colocar en el hijo los valores faltantes de izquierda a derecha, de acuerdo a la secuencia de P2.
5. Repetir los pasos del 1 al 4, pero tomando ahora la secuencia de P2 para el hijo 2.
