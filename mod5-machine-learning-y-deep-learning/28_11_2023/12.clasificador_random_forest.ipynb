## Random Forest (Bosques Aleatorios)

Uno de los problemas con la creacin de un rbol de decisin es que si la profundidad es suficiente, el rbol tiende a "memorizar" las soluciones en vez de generalizar el aprendizaje. Es decir, **tendremos un modelo sobreajustado (overfitting)**.


**La solucin para evitar esto es crear muchos rboles y que trabajen en conjunto.**


**Random Forest es un modelo de aprendizaje supervisado** que, como su nombre lo indica, **consiste en una gran cantidad de rboles de decisin (Decision Trees) individuales que operan como un conjunto**. Cada rbol individual en el bosque aleatorio retorna una prediccin de clase y la clase con ms votos (moda) se convierte en la prediccin de nuestro modelo.




**Por qu se llama Random?**
- Cuenta con una **"doble aleatoriedad"**:
    - La primera es para la seleccin del **valor de "k" (nmero de atributos) para cada rbol**.
    - La segunda es para elegir la **cantidad de patrones (elementos) a utilizar para entrenar** a cada rbol del "bosque".


La razn por la que el modelo de **Random Forest** funciona muy bien es:
Una gran cantidad de modelos **relativamente no correlacionados** (rboles) que funcionan como un grupo superar a cualquiera de los modelos constituyentes individuales.

La **baja correlacin** entre modelos es la clave, los modelos no correlacionados pueden producir predicciones en conjunto que son ms precisas que cualquiera de las predicciones individuales.

La razn de este efecto es que los rboles se "protegen" entre s de sus errores individuales (siempre y cuando no se equivoquen constantemente en la misma direccin).

Si bien algunos rboles pueden retornar predicciones errneas, muchos otros rboles retornaran predicciones correctas, por lo que, como grupo, los rboles pueden retornar predicciones "mas acertadas".

_**Documentacin:** https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html_

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import datasets

# Normalizacion
from sklearn.preprocessing import MinMaxScaler

# Train, Test
from sklearn.model_selection import train_test_split

# Metricas
from sklearn.metrics import jaccard_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score

from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### Procesamiento

```python
# Normalizacin de datos

x_scaler = MinMaxScaler()
X = x_scaler.fit_transform(X)

X
```

### Train, Test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

print(f"X_train: {X_train.shape}, y_train: {y_train.shape}")
print(f"X_test: {X_test.shape},  y_test: {y_test.shape}")
```

### Modelo

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

### Predicciones

```python
yhat = model.predict(X_test)

yhat
```

```python
print("Jaccard Index:", jaccard_score(y_test, yhat, average = "macro"))
print("Accuracy:"     , accuracy_score(y_test, yhat))
print("Precisin:"    , precision_score(y_test, yhat, average = "macro"))
print("Sensibilidad:" , recall_score(y_test, yhat, average = "macro"))
print("F1-score:"     , f1_score(y_test, yhat, average = "macro"))
```

### Confusion Matrix

```python
confusion_matrix(y_test, yhat, labels = [0, 1, 2])
```

### Classification Report

```python
print(classification_report(y_test, yhat, digits = 3))
```

### Parmetros del Modelo (Tuning)

#### Parametros de RandomForestClassifier()

- **`n_estimators`**: int, **`default`** = 100
    - Numero de arboles creados para generar el bosque.

___

- **`criterion`**: {"gini", "entropy"}, **default** = "gini"
    - Es la funcin para medir la calidad de una divisin/split. Los criterios admitidos son "gini" para la impureza de Gini y "entropa" para la ganancia de informacin.

___

- **`max_depth`**: int, **default** = None
    - La profundidad mxima del rbol. Si es **None** los nodos se expanden hasta que todas las hojas sean puras o hasta que todas las hojas contengan menos de **`min_samples_split`** elementos.
        
___

- **`min_samples_split`**: int or float, **default** = 2
    - El nmero mnimo de muestras requeridas para dividir un nodo interno.
    - Si es int, considera **`min_samples_split`** como el nmero mnimo.
    - Si es float, entonces **`min_samples_split`** es una fraccin y **`ceil(min_samples_split * n_samples)`** es el nmero mnimo de muestras para cada divisin/split.
        
___

- **`min_samples_leaf`**: int or float, **`default`** = 1
    - El nmero mnimo de muestras requeridas para llegar a nodo hoja. Es un punto de divisin a cualquier profundidad del arbol  solo se considerar si deja al menos **`min_samples_leaf`** muestras de entrenamiento en cada una de las ramas izquierda y derecha.
    - Si es int, considera **`min_samples_leaf`** como el nmero mnimo.
    - Si es float, entonces **`min_samples_split`** es una fraccin y **`ceil(min_samples_leaf * n_samples)`** es el nmero mnimo de muestras para cada nodo.
        
___

- **`min_weight_fraction_leaf`**: float, **`default`** = 0.0
    - La fraccin ponderada mnima de la suma total de pesos (de todas las muestras de entrada) requerida para estar en un nodo hoja. Las muestras tienen el mismo peso cuando no se proporciona **`sample_weight`**.
            
___

- **`max_features`**: int, float or {"auto", "sqrt", "log2"}, **`default`** = None

    - El nmero de caractersticas (atributos) a considerar al buscar la mejor divisin:
    - Si es int, considera las funciones **`max_features`** en cada divisin.
    - Si es float, **`max_features`** es una fraccin y se consideran **`int(max_features * n_features)`** caractersticas (atributos) en cada divisin.
    - Si es "auto", entonces **`max_features = sqrt(n_features)`**.
    - Si es "sqrt", entonces **`max_features = sqrt(n_features)`**.
    - Si es "log2", entonces **`max_features = log2(n_features)`**.
    - Si None, entonces **`max_features = n_features`**.
  
___

- **`random_state`**: int, **`default`** = None
    - Controla la aleatoriedad del estimador. Las caractersticas (atributos) siempre se permutan aleatoriamente en cada divisin, incluso si el divisor est configurado como "best".
    - Cuando **`max_features < n_features`**, el algoritmo seleccionar **`max_features`** al azar en cada divisin antes de encontrar la mejor divisin entre ellas. Pero la divisin mejor encontrada puede variar entre diferentes ejecuciones, incluso si **`max_features = n_features`**.
    - Ese es el caso, si la mejora del criterio es idntica para varias divisiones y una divisin debe seleccionarse al azar.
    - Para obtener un comportamiento determinista durante el ajuste, **`random_state`** debe fijarse en un nmero entero.
  
___

- **`max_leaf_nodes`**: int, **`default`** = None
    - Este parmetro hace "crecer" el rbol ya que aumenta el nmero de nodos hoja.
    - Se define como "mejor nodo" como una reduccin relativa de la impureza.
    - Si es None, entonces el modelo tendr un nmero ilimitado de nodos hoja.
      
___

- **`min_impurity_decrease`**: float, **`default`** = 0.0
    - Un nodo se dividir si esta divisin induce una disminucin de la impureza mayor o igual a este valor.
         
___

- **`bootstrap`**: bool, **`default`** = True
    - Si se utiliza "bootstrap" al construir rboles.
    - Si es False, se usa todo el conjunto de datos para construir cada rbol.
         
___

- **`n_jobs`**: int, **`default`** = None
    - El nmero de "trabajos" a ejecutar en paralelo. **`fit`**, **`predict`**, **`decision_path`** y **`apply`** estn paralelizados sobre los rboles.
    - None significa 1 procesador para trabajo en paralelo.
    - -1 significa usar todos los procesadores.
             
___


- **`class_weight`**: dict, list of dict or "balanced", **`default`** = None
    - Pesos asociados a cada clases en la forma **`{class_label : weight}`**.
    - Si es None, se supone que todas las clases tienen peso uno.
    - Para problemas de clasificacin mltiple, se puede proporcionar una lista de diccionarios en el mismo orden que la columna y.
     
___

- **`ccp_alpha`**: non-negative float, **`default`** = 0.0
    - Parmetro de complejidad utilizado para **Minimal Cost-Complexity Pruning**.
    - Se elegir el subrbol con la mayor complejidad de costos que sea menor que **`ccp_alpha`**. De forma predeterminada, no se realiza ninguna poda.

```python
# Dataset del titanic preprocesado en clase

titanic = pd.read_csv("../Data/titanic_preprocesamiento.csv")

X = titanic.drop(["Fare-Binning", "Age-Binning", "Survived"], axis = 1)
y = titanic["Survived"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

print(f"X_train: {X_train.shape}, y_train: {y_train.shape}")
print(f"X_test: {X_test.shape},  y_test: {y_test.shape}")
```

```python
# Prueba con RandomForestClassifier() con parmetros por defecto

model = RandomForestClassifier()

# Entrenamiento
model.fit(X_train, y_train)

# Predicciones
yhat_train = model.predict(X_train)
yhat = model.predict(X_test)

# Mtricas
print("Jaccard Index Train:", jaccard_score(y_train, yhat_train, average = "macro"))
print("Accuracy Train:"     , accuracy_score(y_train, yhat_train))
print("Precisin Train:"    , precision_score(y_train, yhat_train, average = "macro"))
print("Sensibilidad Train:" , recall_score(y_train, yhat_train, average = "macro"))
print("F1-score Train:"     , f1_score(y_train, yhat_train, average = "macro"))
print("ROC AUC Train:"      , roc_auc_score(y_train, yhat_train))
print("---------------------------------------------------------------------------")
print("Jaccard Index:", jaccard_score(y_test, yhat, average = "macro"))
print("Accuracy:"     , accuracy_score(y_test, yhat))
print("Precisin:"    , precision_score(y_test, yhat, average = "macro"))
print("Sensibilidad:" , recall_score(y_test, yhat, average = "macro"))
print("F1-score:"     , f1_score(y_test, yhat, average = "macro"))
print("ROC AUC:"      , roc_auc_score(y_test, yhat))
```

```python
# Prueba con DecisionTreeClassifier() usando diferentes parmetros

model = RandomForestClassifier(n_estimators = 200, criterion = "entropy",
                               max_depth = 20, min_samples_split = 15, min_samples_leaf = 8,
                               max_features = "sqrt", random_state = 42, max_leaf_nodes = 40, bootstrap=False)
# Entrenamiento
model.fit(X_train, y_train)

# Predicciones
yhat_train = model.predict(X_train)
yhat = model.predict(X_test)

# Mtricas
print("Jaccard Index Train:", jaccard_score(y_train, yhat_train, average = "macro"))
print("Accuracy Train:"     , accuracy_score(y_train, yhat_train))
print("Precisin Train:"    , precision_score(y_train, yhat_train, average = "macro"))
print("Sensibilidad Train:" , recall_score(y_train, yhat_train, average = "macro"))
print("F1-score Train:"     , f1_score(y_train, yhat_train, average = "macro"))
print("ROC AUC Train:"      , roc_auc_score(y_train, yhat_train))
print("---------------------------------------------------------------------------")
print("Jaccard Index:", jaccard_score(y_test, yhat, average = "macro"))
print("Accuracy:"     , accuracy_score(y_test, yhat))
print("Precisin:"    , precision_score(y_test, yhat, average = "macro"))
print("Sensibilidad:" , recall_score(y_test, yhat, average = "macro"))
print("F1-score:"     , f1_score(y_test, yhat, average = "macro"))
print("ROC AUC:"      , roc_auc_score(y_test, yhat))
```

**Ventajas:**
- Es capaz de dar buenos resultados sin ajuste de parmetros.
- Al utilizar mltiples rboles se reduce considerablemente el riesgo de overfitting.
- Se mantiene estable con nuevas muestras puesto que al utilizar cientos de rboles sigue prevaleciendo la moda de sus votaciones.


**Deventajas:**
- En algunos casos el clasificador Random Forest tambin puede caer en overfitting.
- Es mucho ms "costo" de crear y ejecutar que un solo DecisionTree.
- Puede requerir mucho tiempo de entrenamiento
- Random Forest no funciona bien con datasets pequeos.
- Es muy difcil poder interpretar los cientos de rboles creados en el bosque, si quisiramos comprender y explicar su comportamiento.
