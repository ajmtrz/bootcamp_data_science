## Forward Propagation

Forward Propagation (propagación directa) es la forma de pasar de la capa de entrada (input) a la capa de salida (output) en la red neuronal.



Para una red neuronal:



### Ejercicio:
- Has un algoritmo de Forward Propagation para la imagen de arriba, es decir, una red neuronal de una capa de entrada de 2, una capa oculta de 2 y una capa de salida de 1.

- Inicializa aleatoriamente los pesos, los valores de x y los bias.
- Define una función de activación.

Usa la imagen de arriba como referencia.

```python
import numpy as np
import math

# Tenemos 6 pesos y 3 biases, uno para cada nodo en la capa oculta, así como para el nodo en la capa de salida.

np.random.seed(42)
weights = np.around(np.random.uniform(size = 6), decimals = 2) # Pesos

np.random.seed(42)
x = np.around(np.random.uniform(size = 2), decimals = 2) # X

np.random.seed(42)
biases = np.around(np.random.uniform(size = 3), decimals = 2) # Biases

print(weights)
print(x)
print(biases)
```

```python
def ReLU(x):
    a = list()
    
    for item in x:
        if item < 0:
            a.append(0)
        else:
            a.append(item)
    return a

def sigmoid(x):
    a = (1 / (1 + math.exp(-x)))
    return a
```

```python
z11 = x[0]*weights[0] + x[1]*weights[1] + biases[0]
z12 = x[0]*weights[2] + x[1]*weights[3] + biases[1]
```

```python
z2 = ReLU([z11])[0]*weights[4] + ReLU([z12])[0]*weights[5] + biases[2]
```

```python
a2 = ReLU([z2])
```

```python
a2 = sigmoid(z2)
```
