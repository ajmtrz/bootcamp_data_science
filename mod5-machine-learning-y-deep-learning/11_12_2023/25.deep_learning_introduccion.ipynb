## Deep Learning

**Deep Learning** (aprendizaje profundo) surge a partir del campo de la inteligencia artificial, el cual se define como la agrupación de técnicas que se basan en modelos de redes neuronales artificiales, cuya gran aplicabilidad se ha mostrado recientemente en multitud de usos por parte de la industria.



**Las redes neuronales artificiales** se inspiran en nuestra comprensión de la biología del cerebro humano y de cómo entendemos que actúan las conexiones neuronales.

La idea detrás de las redes neuronales, es bastante simple: imitar cómo funciona el cerebro humano, en específico las conexiones neuronales.



**Las redes neuronales artificiales se organizan como un gráfico dirigido, o matriz compleja de neuronas.**



**La neurona es una máquina que toma una entrada y como resultado de esta entrada genera una salida procesada. En términos matemáticos, cada red de neuronas representa una función no lineal de sus entradas.**


**Cuando una red neuronal comienza a incrementar el número de capas, la estructura de la red comienza a tener más neuronas y más rutas para pasar de una entrada a una salida final por lo que obtiene la capacidad de aprender representaciones más complejas de datos.** 


Respecto a la profundidad de las redes neuronales, esta se debe de equilibrar; debido a que tienen un enorme potencial y poder de aprendizaje; pero, necesita cantidades irracionales de datos y potencial informático para aprender.


Existen muchas maneras en que las neuronas y sus capas se pueden organizar; A cada uno de estos arreglos se les conoce como arquitecturas de redes neuronales. Dependiendo de la tarea en cuestión, se utilizan diferentes arquitecturas para manejar el desafío.
